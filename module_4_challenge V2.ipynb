## Part 1: Explore the Data

Import the data and use Pandas to learn more about the dataset.

```python
import pandas as pd

df = pd.read_csv('Resources/client_dataset.csv')

df.head()
```

```python
# View the column names in the data
df.columns


```

```python
# Add a total price column to the data
df['subtotal'] = df['unit_price'] * df['qty']

shipping_rate = 5  # Example shipping rate
df['shipping_price'] = shipping_rate  # This can be more complex based on weight or other factors.

sales_tax_rate = 0.0925  # 9.25%
df['total_price'] = df['subtotal'] + df['shipping_price']
df['total_price'] += df['total_price'] * sales_tax_rate  # Adding the sales tax.

df.head()



```

```python
# Gather basic statistics
stats = df.describe()
print("Basic Statistics:\n", stats)

# View column names
columns = df.columns
print("\nColumn Names:\n", columns)

# Check data types
data_types = df.dtypes
print("\nData Types:\n", data_types)



```

```python
# Remove rows with missing values and round percentages to 2 decimal places
df = df.dropna()

# List of columns to round
columns_to_round = ['unit_price', 'unit_cost', 'subtotal', 'total_price']

# Round these columns to 2 decimal places
df[columns_to_round] = df[columns_to_round].round(2)

df.head()




```

```python

# Group by 'category' and sum 'total_price', then sort the results
top_categories_sales = df.groupby('category')['total_price'].sum().sort_values(ascending=False).head(3)

# Convert the series to a DataFrame for better formatting
top_categories_sales_df = top_categories_sales.reset_index()

# Format the 'total_price' column as dollars
top_categories_sales_df['total_price'] = top_categories_sales_df['total_price'].apply(lambda x: "${:,.2f}".format(x))

# Display the results
print("Top 3 Categories with Highest Total Sales:")
print(top_categories_sales_df)



```

```python
# For the category with the most entries, which subcategory had the most entries?
# Identify the category with the most entries
top_category = df['category'].value_counts().idxmax()

# Filter the DataFrame for this category
top_category_df = df[df['category'] == top_category]

# Find the subcategory within this category with the most entries
top_subcategory = top_category_df['subcategory'].value_counts().idxmax()

print(f"The subcategory with the most entries in '{top_category}' category is: '{top_subcategory}'")




```

```python
# Which five clients had the most entries in the data?
# Output a df with the client id and the number of entries
# Count the number of entries for each client
client_entries = df['client_id'].value_counts().head(5)

# Convert the series to a DataFrame
top_clients_df = client_entries.reset_index()

# Rename the columns for clarity
top_clients_df.columns = ['client_id', 'number_of_entries']

print("Top 5 Clients by Number of Entries:")
print(top_clients_df)






```

```python
# Store the client ids of those top 5 clients in a list.
# Use that list to filter the original data
# Extracting client IDs and storing them in a list
top_clients_ids = top_clients_df['client_id'].tolist()

# Filtering the DataFrame for entries belonging to the top clients
top_clients_data = df[df['client_id'].isin(top_clients_ids)]

print("Data for the Top 5 Clients:")
print(top_clients_data)


```

## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price and the qty
# Calculate the subtotal for each line
df['subtotal'] = df['unit_price'] * df['qty']

# Display the DataFrame
df.head()




```

```python

# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
# Use the weight column to calculate this.
# Define a function to calculate shipping price based on weight
def calculate_shipping(weight):
    if weight > 50:
        return weight * 7  # $7 per pound for orders over 50 pounds
    else:
        return weight * 10  # $10 per pound for orders 50 pounds or under

# Apply the function to the 'unit_weight' column to create a new 'shipping_price' column
df['shipping_price'] = df['unit_weight'].apply(calculate_shipping)

# Display the DataFrame
df.head()







```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
# Calculate the total price before tax
df['total_price_before_tax'] = df['subtotal'] + df['shipping_price']

# Define the sales tax rate
sales_tax_rate = 0.0925  # 9.25%

# Calculate the final total price including sales tax
df['total_price'] = df['total_price_before_tax'] * (1 + sales_tax_rate)

# Display the DataFrame
df.head()




```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
# Calculate the line cost for each transaction
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']

# Display the DataFrame
df.head()




```

```python
# Create a column for the profit of
# each line using line cost and line price
# Calculate the profit for each line
df['profit'] = df['total_price'] - df['line_cost']

# Display the DataFrame
df.head()




```

```python
# Update the top_clients_df to include shipping price and profit
# Group the main DataFrame by 'client_id' and calculate the total shipping price and profit
# Group the main DataFrame by 'client_id' and calculate the total shipping price and profit
grouped_data = df.groupby('client_id')[['shipping_price', 'profit']].sum()

# Update the top_clients_df to include this data
top_clients_df = top_clients_df.merge(grouped_data, on='client_id', how='left')

# Display the updated DataFrame
print("Top Clients with Shipping Price and Profit:")
print(top_clients_df)





```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# Ensure top_clients_ids contains the IDs of the top 5 clients
top_clients_ids = [33615, 66037, 46820, 38378, 24741]

# Filter the DataFrame for the top 5 clients
top_clients_data = df[df['client_id'].isin(top_clients_ids)]

# Group by client_id and sum the total_price for each client
top_clients_spending = top_clients_data.groupby('client_id')['total_price'].sum()

# Display the total spending for each of the top 5 clients
for client_id in top_clients_ids:
    spending = top_clients_spending.loc[client_id]
    print(f"Client ID {client_id} spent a total of: ${spending:.2f}")





```

```python
# Summarize the data for the top 5 clients and display as dollars
# Aggregating the required data for each of the top clients
top_clients_aggregated = df[df['client_id'].isin(top_clients_ids)].groupby('client_id').agg(
    total_spending=('total_price', 'sum'),
    total_shipping=('shipping_price', 'sum'),
    total_profit=('profit', 'sum')
)

# Formatting the values as dollars
top_clients_aggregated['total_spending'] = top_clients_aggregated['total_spending'].apply(lambda x: f"${x:,.2f}")
top_clients_aggregated['total_shipping'] = top_clients_aggregated['total_shipping'].apply(lambda x: f"${x:,.2f}")
top_clients_aggregated['total_profit'] = top_clients_aggregated['total_profit'].apply(lambda x: f"${x:,.2f}")

# Display the summary DataFrame
top_clients_aggregated




```
